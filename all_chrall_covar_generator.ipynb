```python
pheno_temp = pheno[['IID_fix', 'IID', 'ORIGPROT' ]]
pheno_temp['IID_fix_origprot'] = pheno_temp['IID_fix'] + "_" + pheno_temp['ORIGPROT']
print(pheno_temp.IID_fix_origprot.nunique() )
pheno_temp = pheno_temp.drop_duplicates(subset='IID_fix_origprot', keep="last")
print(pheno_temp.shape)
```

```python

pheno_temp = pheno_temp.drop(['IID_fix_origprot'], axis = 1)
print(pheno_temp.shape)
print(pheno_temp.head())
```

```python
ADNIMERGE_NEW = ADNIMERGE.merge(pheno_temp, how='left', left_on=['PTID','ORIGPROT'], right_on = ['IID_fix','ORIGPROT'] )
print(ADNIMERGE_NEW.shape)
print(ADNIMERGE_NEW.head())
print(ADNIMERGE_NEW.describe())
```

```python
with open('all_chrall_filtered_callRate.fam') as f:
    fam = f.readlines()
fam2 = []
for i in range(len(fam)):
    fam[i] = fam[i].split()
    PTID = '_'.join(fam[i][1].split('_')[1:])
#     print(PTID)
    try:
        CDRSB = ADNIMERGE[['PTID', 'Years_bl', 'CDRSB']][ADNIMERGE['PTID'] == PTID].dropna().sort_values(by='Years_bl').iloc[-1]['CDRSB']
        fam[i][-1] = str(CDRSB)
        fam2 += [fam[i]]
    except Exception as e:
        print(e)
#         raise
fam2
```

```python
fam2 = [' '.join(x) for x in fam2]
```

```python
with open('all_chrall_filtered_callRate.fam', 'w') as f:
    f.write('\n'.join(fam2))
```
